```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")


penguins = sns.load_dataset("penguins") # numricas + categricas (especie, isla, etc.)


pearson_corr = penguins["bill_length_mm"].corr(penguins["flipper_length_mm"]) # Calcular correlacin de Pearson

plt.figure(figsize=(10,5))
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='island', style='sex', alpha=1, s=50)
sns.regplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', color='red', scatter=False)
plt.title('Bill Length VS Flipper Length')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Flipper Length (mm)')
plt.figtext(.30, .80,'Pearson Correlation: 0.68', fontsize=12, color='red')
```

```python
penguins_num = penguins.select_dtypes(include='number')
corr_penguins = penguins_num.corr()
display(corr_penguins)

plt.figure(figsize=(10,5))
sns.heatmap(corr_penguins, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

```
